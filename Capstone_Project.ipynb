```python
# 1. Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load the dataset
df = pd.read_csv('IMDB Dataset.csv')
```

```python
# 3. Display basic information and check for null values
print("--- Dataset Information ---")
df.info()
print("\n--- First 5 Rows ---")
print(df.head())
print("-" * 30)
```

```python
# 4. Task 6: Movie with the Highest Votes
print("\n--- Task 6: Movie with the Highest Votes ---")
highest_votes_movie = df.loc[df['No_of_Votes'].idxmax()]
print(highest_votes_movie[['Series_Title', 'Genre', 'Director', 'No_of_Votes', 'IMDB_Rating']])
print("-" * 30)
```

```python
# 5. Task 7: Create boxplots
print("\n--- Task 7: Generating Boxplots ---")
plt.figure(figsize=(12, 5))
plt.suptitle('Boxplots for IMDB Rating and Runtime')
plt.subplot(1, 2, 1)
sns.boxplot(y=df['IMDB_Rating'])
plt.title('Boxplot of IMDB Rating')
plt.subplot(1, 2, 2)
# The 'Runtime' column is text ('142 min'), so we must clean it first.
# df['Runtime_Clean'] = df['Runtime'].str.replace(' min', '').astype(int) # This line caused an error as 'Runtime' is already int
sns.boxplot(y=df['Runtime']) # Use the original 'Runtime' column
plt.title('Boxplot of Runtime (Minutes)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
print("-" * 30)
```

```python
# 6. Task 8: Check the relationship between Rating and Runtime
print("\n--- Task 8: Relationship between Rating and Runtime ---")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Runtime', y='IMDB_Rating', data=df) # Use the original 'Runtime' column
plt.title('Relationship: IMDB Rating vs. Runtime')
plt.xlabel('Runtime (Minutes)')
plt.ylabel('IMDB Rating')
plt.grid(True)
plt.show()
print("-" * 30)
```

```python
# 7. Task 9: Check the distribution of features
print("\n--- Task 9: Checking Feature Distributions ---")
plt.figure(figsize=(14, 6))
plt.suptitle('Distribution of IMDB Rating and Runtime')
plt.subplot(1, 2, 1)
sns.histplot(df['IMDB_Rating'], kde=True, bins=20)
plt.title('Distribution of IMDB Rating')
plt.subplot(1, 2, 2)
sns.histplot(df['Runtime'], kde=True, bins=30) # Use the original 'Runtime' column
plt.title('Distribution of Runtime (Minutes)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
print("-" * 30)
```

```python
# 8. Task 10: Create a count plot of movies by genre
print("\n--- Task 10: Count Plot of Movies by Genre ---")
plt.figure(figsize=(10, 8))
# This task requires splitting the 'Genre' string to get the primary genre.
df['Primary Genre'] = df['Genre'].apply(lambda x: x.split(',')[0])
# Get the top 10 most frequent primary genres for a cleaner plot.
top_10_genres = df['Primary Genre'].value_counts().nlargest(10).index
# Filter the DataFrame to include only these top 10 genres.
df_top_genres = df[df['Primary Genre'].isin(top_10_genres)]
sns.countplot(y='Primary Genre', data=df_top_genres, order=top_10_genres)
plt.title('Top 10 Number of Movies by Primary Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
print("-" * 30)
```
